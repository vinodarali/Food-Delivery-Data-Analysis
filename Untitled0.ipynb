```python
import pandas as pd
import json
import sqlite3

orders = pd.read_csv('orders.csv')

with open('users.json', 'r') as f:
    users_data = json.load(f)
users = pd.DataFrame(users_data)

conn = sqlite3.connect(':memory:')
with open('restaurants.sql', 'r') as f:
    sql_script = f.read()
conn.executescript(sql_script)
restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
conn.close()
```

```python
merged_data = pd.merge(orders, users, on='user_id', how='left')

final_df = pd.merge(merged_data, restaurants, on='restaurant_id', how='left')

print(f"Final dataset rows: {len(final_df)}")
final_df.head()
```

```python
gold_orders = final_df[final_df['membership'] == 'Gold'].shape[0]

hyd_rev = round(final_df[final_df['city'] == 'Hyderabad']['total_amount'].sum())

gold_aov = round(final_df[final_df['membership'] == 'Gold']['total_amount'].mean(), 2)

print(f"Gold Orders: {gold_orders}")
print(f"Hyderabad Revenue: {hyd_rev}")
print(f"Gold AOV: {gold_aov}")

final_df.to_csv('final_food_delivery_dataset.csv', index=False)
```
